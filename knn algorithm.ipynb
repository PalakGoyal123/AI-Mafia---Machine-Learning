```python
# start with knn code
# step-1 is to find k nearest neighbours
#based on distance we can find the value of k
```

```python
# distance formula
def distance(x1,x2):
    return np.sqrt(sum(x1-x2)**2)
    
    
```

```python
def knn(x,y,querypoint,k = 5):
    #pick k nearest neighbour for every point in the x
    val =[]
    for i in range(x.shape[0]):
        #compute distance
        d = distance(querypoint,x[i])
        val.append((d,y[i]))
    print(val)
    #2.sort the array and find the k nearest points
    val = sorted(val)
    #3.take first k points
    val =val[:k]
    #majority vote
    val = np.array(val)
    new_values = np.unique(val[:,1],return_counts = True)
    print(new_values)
    
    index = new_values[1].argmax()
    pred = new_values[0][index]
    return pred
            
            
        
    
    
    
```
